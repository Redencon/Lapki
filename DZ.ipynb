```python
data2 = pd.read_csv('DZ2.csv', sep=';')
plt.figure(figsize=(12, 8))
def comma_to_dot(s: str) -> str:
    if type(s) != str:
        return s
    a = ''
    for c in s:
        if c == ',':
            a += '.'
        else:
            a += c
    return a

data2 = data2.applymap(comma_to_dot).astype('float64')
data2['Uv_real, V'] = data2['Uv, V'] - data2['Ui, V']
data2['I, mA'] = data2['Ui, V'] / 9.1
ax = plt.axes()
sns.scatterplot(data = data2, x = 'Uv_real, V', y = 'I, mA', ax=ax, hue='Uvch, V', palette='winter', marker='o')
sns.scatterplot(data = data, x = 'Uv_real, V', y = 'I, mA', ax=ax, hue='Uvch, V', palette='autumn', marker='o')

ax.set_xlabel('Напряжение между электродами, В')
ax.set_ylabel('Ток между электродами')
ax.set_title('ВАХ двойного зонда')

from matplotlib.ticker import MultipleLocator
plt.grid(which='major', color='#323232', linestyle='-')
plt.grid(which='minor', color='#a4a4a4', linestyle=':')
mlx = MultipleLocator(2)
ax.xaxis.set_minor_locator(mlx)
mly = MultipleLocator(0.005)
ax.yaxis.set_minor_locator(mly)
# data.to_csv('DZ.csv')
```

```python
def f(U, dU, T, I0, a, b):
    return I0 * np.tanh(1.6e-19*(U - dU)/(2*1.38e-23*T)) + a*U + b

Us = (203, 250, 300, 350, 400)
columns = ('Uvch', 'dU', 'T', 'I0, mA', 'a', 'b')
def hehehe(data, p0, func=f, cols = columns):
    plt.figure(figsize=(12, 8))
    ax = plt.axes()
    popt = []
    pcov = []
    for Uvch in p0:
        popti, pcovi = curve_fit(func, data[data['Uvch, V'] == Uvch]['Uv, V'], data[data['Uvch, V'] == Uvch]['I, mA'], maxfev=5000**2, p0=p0[Uvch])
        popt.append([Uvch] + list(popti))
        pcov.append([Uvch] + list(np.sqrt(pcovi.diagonal())))
        xs = np.linspace(-80, 80, 400)
        ys = func(xs, *popti)
        ax.plot(xs, ys)
        ax.scatter(data[data['Uvch, V'] == Uvch]['Uv, V'], data[data['Uvch, V'] == Uvch]['I, mA'])
    popts = pd.DataFrame(popt, columns=cols)
    pcovs = pd.DataFrame(pcov, columns=cols)
    return popts, pcovs, ax

popts, pcovs, ax = hehehe(data, {Uvch: (0, 1e5, 0, 0, 0) for Uvch in Us})
# popts['I0, A'] = popts['I0'] / 9.1e3
popts['n'] = 6.02e22 * popts['I0, mA']*1e-3 / np.sqrt(popts['T'])
popts['alpha'] = popts['n']/1.45e21
popts['d_alpha'] = np.sqrt(0.5*(pcovs['T'] / popts['T'])**2 + (pcovs['I0, mA']/popts['I0, mA'])) * popts['alpha']
popts['d_T'] = pcovs['T']
popts['d_n'] = np.sqrt(0.5*(pcovs['T'] / popts['T'])**2 + (pcovs['I0, mA']/popts['I0, mA'])) * popts['n']

ax.set_xlabel('Напряжение на зонде, В')
ax.set_ylabel('Ток на зонде, мА')
ax.set_title('ВАХ двойного зонда при давлении 60 мкбар')

from matplotlib.ticker import MultipleLocator
plt.grid(which='major', color='#323232', linestyle='-')
plt.grid(which='minor', color='#a4a4a4', linestyle=':')
mlx = MultipleLocator(2)
ax.xaxis.set_minor_locator(mlx)
mly = MultipleLocator(0.005)
ax.yaxis.set_minor_locator(mly)
# data.to_csv('DZ.csv')
```
