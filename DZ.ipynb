```python
from scipy.optimize import curve_fit
```

```python
def f(U, dU, T, I0, a, b):
    return I0 * np.tanh(1.6e-19*(U - dU)/(2*1.38e-23*T)) + a*U + b

Us = (203, 250, 300, 350, 400)
def hehehe(data, p0):
    plt.figure(figsize=(12, 8))
    popt = []
    pcov = []
    for Uvch in p0:
        popti, pcovi = curve_fit(f, data[data['Uvch, V'] == Uvch]['Uv, V'], data[data['Uvch, V'] == Uvch]['Ui, V'], maxfev=5000**2, p0=p0[Uvch])
        popt.append([Uvch] + list(popti))
        pcov.append([Uvch] + list(np.sqrt(pcovi.diagonal())))
        xs = np.linspace(-80, 80, 400)
        ys = f(xs, *popti)
        plt.plot(xs, ys)
        plt.scatter(data[data['Uvch, V'] == Uvch]['Uv, V'], data[data['Uvch, V'] == Uvch]['Ui, V'])
    popts = pd.DataFrame(popt, columns=('Uvch', 'dU', 'T', 'I0', 'a', 'b'))
    pcovs = pd.DataFrame(pcov, columns=('Uvch', 'dU', 'T', 'I0', 'a', 'b'))
    return popts, pcovs
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(data=popts, x='Uvch', y='T')
plt.errorbar(x = popts['Uvch'], y=popts['T'], yerr=pcovs['T'], linestyle='')
```

```python
data2 = pd.read_csv('DZ2.csv', sep=';')
plt.figure(figsize=(12, 8))
def comma_to_dot(s: str) -> str:
    if type(s) != str:
        return s
    a = ''
    for c in s:
        if c == ',':
            a += '.'
        else:
            a += c
    return a

data2 = data2.applymap(comma_to_dot).astype('float32')
ax = plt.axes()
sns.scatterplot(data = data2, x = 'Uv, V', y = 'Ui, V', ax=ax, hue='Uvch, V', palette='viridis', marker='^')
# sns.scatterplot(data = data, x = 'Uv, V', y = 'Ui, V', ax=ax, hue='Uvch, V', palette='viridis', marker='^')
```
